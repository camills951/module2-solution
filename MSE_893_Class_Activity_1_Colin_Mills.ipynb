```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/data_with_outliers.csv') # Replace 'data_file.csv' with the actual file path.
data.head()
```

```python
mean = data.mean()
print('mean =' , mean)
median = data.median()
print('median =' , median)
mode = data.mode().iloc[0]
print('mode =' , mode)
std_dev = data.std()
print('standard deviation =' , std_dev)
```

```python
# Plotting X value over the sequence of records
plt.plot(data.iloc[1:,0])
plt.xlabel('Time')
plt.ylabel('X Value')
plt.title('Data Trend X - Fig 1')
plt.show()
```

```python
# Plotting Y value over the sequence of records
plt.plot(data.iloc[1:,1])
plt.xlabel('Time')
plt.ylabel('Y Value')
plt.title('Data Trend Y - Fig 2')
plt.show()
```

```python
ax = data['X'].plot(kind='hist', bins=20, title='Histogram of X - Fig 3',log=True)
ax.set_xlabel('X')
```

```python
ax = data['Y'].plot(kind='hist', bins=20, title='Histogram of Y - Fig 4',log=True, color = 'red')
ax.set_xlabel('Y')
```

```python
correlation = data['X'].corr(data['Y'])
print(correlation)
```

```python
window_size = 10 # Adjust the window size as desired
moving_average = data.rolling(window_size).mean()
plt.plot(moving_average)
```

Conclusion and Findings

- There is a **weak negative correlation** between the X and Y features. So a change in the value of one of these features will have little or no effect on the values of the other feature.
- The **data trend** for both X and Y over the sequence of time records **does not show any obvious significant patterns** (Figures 1 and 2)
- The values of **Y** have a **higher frequency** in the higher number range (**80 to 140**, Figure 4) versus the values of **X** where they are generally **lower** and **between 30 and 70** (Figure 3).
- In the moving average graph, the **X** feature (represented in blue) shows a **slight increase** in value as the sequence of time records progresses. The **Y** feature (represented in orange) shows a **slight decrease** as the time records progresses. The window averaging helps demonstrate these patterns more obviously than the data trending with averages in Figures 1 and 2.
